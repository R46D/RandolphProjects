```python

import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(ChromeDriverManager().install())



# Scroll page depending our requirement
def scroll_page():
    for x in range(0):
        html = driver.find_element(By.TAG_NAME,'html')
        html.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)


def scrap_videos():
    global cn, subs,titles,urls,viewss,post_dates
    scroll_page()

    cx = '//*[@id="channel-name"]'
    sx = '//*[@id="subscriber-count"]'
    vc = "style-scope ytd-rich-grid-media"
    vx = './/*[@id="metadata-line"]/span[1]'
    pdx = './/*[@id="metadata-line"]/span[2]'
    url_id= 'video-title-link'

    tx = './/*[@id="video-title"]'

    # Scrap Channel Name
    try:
        cn = driver.find_element(By.XPATH,cx).text
    except (Exception,):
        pass

    # Scrap Number of Subscribers
    try:
        subs = driver.find_element(By.XPATH,sx).text
    except (Exception,):
        pass

    # Reassign variable to recalculate all videos
    videos = driver.find_elements(By.CLASS_NAME,vc)

    # Loop through all videos
    for video in videos:

        # grab title if available
        try:
            titles = video.find_element(By.XPATH,tx).text
        except (Exception,):
            pass

        # grab url if available
        try:
            urls = video.find_element(By.ID,url_id).get_attribute("href")
        except (Exception,):
            pass

        # grab views if available
        try:
            viewss = video.find_element(By.XPATH,vx).text
        except (Exception,):
            pass

        # grab post date if available
        try:
            post_dates = video.find_element(By.XPATH,pdx).text
        except (Exception,):
            pass

        video_items = {
            "channel_name": cn,
            "subscribers": subs,
            "title": titles,
            "views": viewss,
            "post_date": post_dates,
            "url": urls,
        }

        vid_list.append(video_items)

    return vid_list


#list of channels to scrape
list_channels = ["https://www.youtube.com/@IGN/videos","https://www.youtube.com/@WatchMojo/videos"]
# empty list to hold video details
vid_list = []


for channel in range(len(list_channels)):
    url = list_channels[channel]
    driver.get(url)

    vid_list = scrap_videos()
    driver.implicitly_wait(10)

# Close Chrome browser
driver.quit()

# create pandas df for video info
df_channel = pd.DataFrame(vid_list)

# export df to csv
df_channel.to_csv("yt_channel_scrap.csv")
```
